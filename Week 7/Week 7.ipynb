## "Exercise 1: Dimension reduction with PCA"

```python
#1.Load the data above into a dataframe.
dataframe = pd.read_csv("censusCrimeClean (1).csv")
display(dataframe)
```

```python
#2.Since PCA only works with numerical data, extract all the data except for the firsty column.
numerical_dataframe = dataframe.drop('communityname', axis = 1)
```

```python
#3.Create and fit PCA with two components with these features.
pca=PCA(n_components=2)
pca.fit(numerical_dataframe)
```

```python
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)
```

```python
components = pca.transform(numerical_dataframe)
print(components)
plt.scatter(components[:,0], components[:,1])
plt.title('Principal Component Analysis')
plt.xlabel('PC_1')
plt.ylabel('PC_2')
```

## "Exercise 2: Multidimensional Scaling"

```python
#1.Load the data from here
wine_data = pd.read_excel("london-borough-profilesV2.xlsx")
display(wine_data)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
# Extract and store numeric features (excluding the first column) and known labels
numeric_features = wine_data.select_dtypes(include=['float64', 'int64'])
features = numeric_features.iloc[:, 1:]  # Assuming the first column is the label column
known_labels = wine_data.iloc[:, 0]

# Step 2: Use k-means to cluster the features into 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Handle non-numeric values by excluding them
features_numeric = features.select_dtypes(include=['float64', 'int64'])

# Fill NaN values with 0
features_numeric = features_numeric.fillna(0)

cluster_labels = kmeans.fit_predict(features_numeric)

# Encode known_labels to numeric type using LabelEncoder
label_encoder = LabelEncoder()
known_labels_encoded = label_encoder.fit_transform(known_labels)

# Step 3: Compare known labels with cluster labels
# Create a confusion matrix
conf_matrix = confusion_matrix(known_labels_encoded, cluster_labels)

# Step 4: Plot wines in a scatterplot using PCA
# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(features_numeric)

# Create a DataFrame for plotting
plot_df = pd.DataFrame({"PC1": pca_result[:, 0], "PC2": pca_result[:, 1], "Known Labels": known_labels_encoded, "Cluster Labels": cluster_labels})

# Scatter plot with known labels
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x="PC1", y="PC2", hue="Known Labels", data=plot_df, palette="viridis", legend="full")
plt.title("Scatter Plot with Known Labels")

# Scatter plot with cluster labels
plt.subplot(1, 2, 2)
sns.scatterplot(x="PC1", y="PC2", hue="Cluster Labels", data=plot_df, palette="viridis", legend="full")
plt.title("Scatter Plot with Cluster Labels")

plt.show()
```

## "Exercise 3: Group wines based on their characteristics"

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the data
url = "wine.csv"  # Replace with the actual URL or local path to the wine data CSV file
wine_data = pd.read_csv(url)

# Extract and store features (excluding the first column) and known labels
features = wine_data.iloc[:, 1:]  # Assuming the first column is the label column
known_labels = wine_data.iloc[:, 0]

# Step 2: Use k-means to cluster the features into 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(features)

# Encode known_labels to numeric type using LabelEncoder
label_encoder = LabelEncoder()
known_labels_encoded = label_encoder.fit_transform(known_labels)

# Step 3: Compare known labels with cluster labels
# Create a confusion matrix
conf_matrix = confusion_matrix(known_labels_encoded, cluster_labels)
print("Confusion Matrix:")
print(conf_matrix)

# Step 4: Plot wines in a scatterplot using PCA
# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(features)

# Create a DataFrame for plotting
plot_df = pd.DataFrame({"PC1": pca_result[:, 0], "PC2": pca_result[:, 1], "Known Labels": known_labels_encoded, "Cluster Labels": cluster_labels})

# Scatter plot with known labels
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x="PC1", y="PC2", hue="Known Labels", data=plot_df, palette="viridis", legend="full")
plt.title("Scatter Plot with Known Labels")

# Scatter plot with cluster labels
plt.subplot(1, 2, 2)
sns.scatterplot(x="PC1", y="PC2", hue="Cluster Labels", data=plot_df, palette="viridis", legend="full")
plt.title("Scatter Plot with Cluster Labels")

plt.show()

# Optional: Investigate feature importance using Decision Tree
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(features, known_labels_encoded)

# Display feature importances
feature_importances = pd.DataFrame({"Feature": features.columns, "Importance": decision_tree.feature_importances_})
feature_importances = feature_importances.sort_values(by="Importance", ascending=False)
print("\nFeature Importance:")
print(feature_importances)
```

## "Exercise 4"

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
census = pd.read_csv("censusCrimeClean (1).csv")
independent = census["medIncome"]
dependent = census["ViolentCrimesPerPop"]
# Initialize KFold with k=5
kf = KFold(n_splits=5, shuffle=True, random_state=42)
foldCount = 0
for train_idxs, test_idxs in kf.split(census):
    print("Run: ", foldCount)
    # Extract the test (unseen) and training subsets
    dependentSubset = dependent.iloc[train_idxs]
    independentSubset = independent.iloc[train_idxs]
    dependentSubsetUnseen = dependent.iloc[test_idxs]
    independentSubsetUnseen = independent.iloc[test_idxs]
    # Build a regression model on the training set
    model = LinearRegression()
    model.fit(independentSubset.values.reshape(-1, 1), dependentSubset)
    # Report the regression model parameters
    print("Intercept:", model.intercept_)
    print("Coefficient:", model.coef_[0])
    # Plot
    plt.scatter(independentSubsetUnseen, dependentSubsetUnseen, color='black', label='Actual')
    plt.plot(independentSubsetUnseen, model.predict(independentSubsetUnseen.values.reshape(-1, 1)), color='blue', linewidth=3, label='Predicted')
    plt.xlabel("Median Income")
    plt.ylabel("Violent Crimes Per Population")
    plt.legend()
    plt.show()
    # Predict values for the test set
    predictions = model.predict(independentSubsetUnseen.values.reshape(-1, 1))
    # Report the residuals
    residuals = dependentSubsetUnseen - predictions
    print("Residuals:", residuals)
    foldCount += 1
```
